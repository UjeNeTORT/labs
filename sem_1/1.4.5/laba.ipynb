## Теоретические сведения

### Вывод уравнения волны на струне 
----

Буквой $T$ обозначим силы натяжения в струне, они должны превышать ее собственный вес, чтобы струна не провисала.

Второй закон Ньютона для вертикального движения элемента струны:

$$ \delta m \ddot{y} = -T_{1} \sin \alpha_{1} + ~T_{2} \sin \alpha_{2}$$

В силу того, что рассматриваемый элемент струны мал, можно считать $ \tg \alpha = \alpha = \frac{\delta y}{\delta x}$, $T_{2} = T_{1} = T$

Поделим уравнение на $\delta x \to 0$

$$ \rho_{l} \ddot{y} = T \frac{\alpha_{2} - \alpha_{1}}{\delta x} \to T \frac{\delta \alpha}{\delta x}$$

$\rho_{l}$ - погонная плотность струны (масса на единицу длины)

Введем величину 
$$ u = \sqrt{\frac{T}{\rho_{l}}}$$

Получим **волновое уравнение**:

$$ \frac{\delta^{2} y}{\delta t^{2}} = u^{2} ~T \frac{\delta^{2} y}{\delta x^{2}} $$

### Бегущие волны
----


Величина $u$, которую мы ввели выше есть _скорость распространения волны_ по струне

Решив дифференциальное уравнение получим сумму двух волн произвольной формы:

$$ y(x, t) = y_{1}(x - ut) + y_{2} (x + ut) $$

Для случая гармонических волн:

$$ y(x, t) = a \cos(\omega t - kx) + b \cos(\omega t + kx) $$

Это выражение представляет собой суперпозицию двух волн бегущих навстречу друг другу со скоростью $u = \frac{\omega}{k} = \nu \lambda $ ($ k = \frac{2\pi}{\lambda}$ - волновое число)

__ОТМЕТИМ, ЧТО СКОРОСТЬ $u$ ЗАВИСИТ ТОЛЬКО ОТ $T$ И $\rho_{l}$!__  
В силу малости амплитуды модуль юнга не учитывается.

Возбуждение стоячих колебаний на струне с закрепленными концами возможно если на длине струны укладывается целое число полуволн.
$$\lambda_{n} = \frac{2L}{n}$$

### Графики зависимости частот от номеров гармоник


```python
def LeastSquares(x_arr, y_arr): # returns [k, b, sigma_k, sigma b]
    v = x_arr
    u = y_arr

    N = len(v)

    mu = np.mean(u) # средее
    mv = np.mean(v)
    mv2 = np.mean([v_i**2 for v_i in v]) # средний квадрат
    mu2 = np.mean([u_i**2 for u_i in u])
    muv = np.mean ([u[i] * v[i] for i in range(len(u))]) # среднее от произведения
    k = (muv - mu * mv) / (mv2 - mv**2)
    b = mu - k * mv

    sigma_k = np.sqrt(1/(N-2) * ( (mu2 - mu**2)/(mv2 - mv**2) - k**2 ) )
    sigma_b = sigma_k * np.sqrt(mv2)

    return (k, b, sigma_k, sigma_b)

```

```python
plt.figure()
plt.ylabel(" nu, Гц")
plt.xlabel("n")
plt.title('Зависимость частот от номеров гармоник')
plt.grid(True, linestyle = '--')
plt.errorbar(nu_T_1['hm_n'], nu_T_1['arr'], fmt = '.k')
plt.errorbar(nu_T_2['hm_n'], nu_T_2['arr'], fmt = '.k')
plt.errorbar(nu_T_3['hm_n'], nu_T_3['arr'], fmt = '.k')
plt.errorbar(nu_T_4['hm_n'], nu_T_4['arr'], fmt = '.k')
plt.errorbar(nu_T_5['hm_n'], nu_T_5['arr'], fmt = '.k')
x = np.array([0, 10])

z = 0  # микро костыль

k1, b1, sigma_k1, z = LeastSquares(nu_T_1['hm_n'], nu_T_1['arr'])
plt.plot(x, k1*x + b1 , color = 'r', linewidth = 1, label = "T = 9.82 H")

k2, b2, sigma_k2, z = LeastSquares(nu_T_2['hm_n'], nu_T_2['arr'])
plt.plot(x, k2*x + b2 , color = 'g', linewidth = 1, label = "T = 14.32 H")

k3, b3, sigma_k3, z = LeastSquares(nu_T_3['hm_n'], nu_T_3['arr'])
plt.plot(x, k3*x + b3 , color = 'b', linewidth = 1, label = "T = 19.24 H")

k4, b4, sigma_k4, z = LeastSquares(nu_T_4['hm_n'], nu_T_4['arr'])
plt.plot(x, k4*x + b4 , color = 'c', linewidth = 1, label = "T = 23.90 H")

k5, b5, sigma_k5, z = LeastSquares(nu_T_5['hm_n'], nu_T_5['arr'])
plt.plot(x, k5*x + b5 , color = 'm', linewidth = 1, label = "T = 28.76 H")

plt.legend()
```

Частота колебаний струны
$$ \nu_{n} = \frac{u}{\lambda_{n}} = \frac{u n}{2L}$$
То есть на графике $\nu(n)$ угловой коэффициент $k = u/2L$, откуда скорость волны $ u = 2Lk $

```python
u1 = 2 * L * k1
u2 = 2 * L * k2
u3 = 2 * L * k3
u4 = 2 * L * k4
u5 = 2 * L * k5
```

$$ \sigma_{u} = \sqrt{4L^{2}\sigma_{k} + 4k^{2} \sigma_{L}}$$

```python
sigma_u1 = math.sqrt(4 * L**2 * sigma_k1 + 4 * k1**2 * sigma_L)
sigma_u2 = math.sqrt(4 * L**2 * sigma_k2 + 4 * k2**2 * sigma_L)
sigma_u3 = math.sqrt(4 * L**2 * sigma_k3 + 4 * k3**2 * sigma_L)
sigma_u4 = math.sqrt(4 * L**2 * sigma_k4 + 4 * k4**2 * sigma_L)
sigma_u5 = math.sqrt(4 * L**2 * sigma_k5 + 4 * k5**2 * sigma_L)
```

### Построение графика $u^{2}(T)$, определение погонной плотности струны $ \rho_{l} $

```python
u_sqr = [u1**2, u2**2, u3**2, u4**2, u5**2]
T     = [nu_T_1["T"], nu_T_2["T"], nu_T_3["T"], nu_T_4["T"], nu_T_5["T"]]
sigma_u_sqr = [2 * u1 * sigma_u1,
               2 * u2 * sigma_u2,
               2 * u3 * sigma_u3,
               2 * u4 * sigma_u4,
               2 * u5 * sigma_u5,]
sigma_T = [0, 0, 0, 0, 0]

kerr, z = 0, 0
```

```python
plt.figure()
plt.ylabel("$u^{2}$, м/c^2")
plt.xlabel("$T$")
plt.title('$u^{2}(T)$')
plt.grid(True, linestyle = '--')
plt.errorbar(T, u_sqr, xerr=sigma_T, yerr = sigma_u_sqr, fmt = ".k")
x = np.array([0, 30])
k, b, kerr, z = LeastSquares(T, u_sqr)
plt.plot(x, k*x + b, "-r", linewidth = 1)
```

$$u= \sqrt{\frac{T}{\rho_{l}}} $$

$$ u^{2} = \frac{T}{\rho_{l}}$$

По графику $u^{2}(T)$ угловой коэффициент $$k = \frac{1}{\rho_{l}}$$

$$ \rho_{l} = \frac{1}{k} $$
$$ \sigma_{\rho} = \sigma_{k} / k^{2} $$


```python
rho_l = 1 / k
sigma_rho_l = kerr / (k**2)
rho_l, sigma_rho_l
```

$$ \rho_{l} = 583.4 \pm 30.5 ~мг/м$$
