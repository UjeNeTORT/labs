```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
questions = [1, 2, 3, 4, 5]
# questions = ["wakes up early this academic year", "has classes on Saturdays", "has classes on Saturdays", "has lab works twice a week", "goes out with friends regularly", "watches pre-recorded lectures"]

answers = [(3, 1), (3, 2), (7, 0), (2, 2), (3, 7)]

ans_now = [i[0] for i in answers]
ans_then = [i[1] for i in answers]
```

```python
X_axis = np.arange(len(questions))
```

```python
plt.figure()
plt.grid(True, linestyle="--")

plt.bar(X_axis + 0.2, ans_then, 0.4, label = '2023')
plt.bar(X_axis - 0.2, ans_now, 0.4, label = '2024')

plt.xticks(X_axis, questions)
plt.xlabel("question")
plt.ylabel("N students")
plt.title("Number of students in each group")

plt.legend()
plt.show()
```
