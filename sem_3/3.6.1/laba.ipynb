# 3.6.1 Спектральный анализ 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def LeastSquares(x_arr, y_arr): # returns [k, b, sigma_k, sigma b]
    v = x_arr
    u = y_arr

    N = len(v)

    mu = np.mean(u) # средее
    mv = np.mean(v)
    mv2 = np.mean([v_i**2 for v_i in v]) # средний квадрат
    mu2 = np.mean([u_i**2 for u_i in u])
    muv = np.mean ([u[i] * v[i] for i in range(len(u))]) # среднее от произведения
    k = (muv - mu * mv) / (mv2 - mv**2)
    b = mu - k * mv

    sigma_k = np.sqrt(1/(N-2) * ( (mu2 - mu**2)/(mv2 - mv**2) - k**2 ) )
    sigma_b = sigma_k * np.sqrt(mv2)

    return (k, b, sigma_k, sigma_b)
```

```python
tau = [50, 25, 20, 16.7, 12.5, 10]
nu  = [51.1, 25, 19.2, 17, 12.5, 10]

k, b, kerr, berr = LeastSquares(tau, nu)
```

```python
plt.figure()
plt.title("$\\Delta \\nu (1/\\tau)$")
plt.xlabel("$1/\\tau \\cdot 10^{3} c^{-1}$")
plt.ylabel("$\\Delta \\nu, кГц$")

plt.grid(True, linestyle="--")
plt.errorbar([i for i in tau], [i for i in nu], fmt=".k")
x = np.array([min(tau), max(tau)])
plt.plot(x, k*x + b, label=f'k = {round(k, 2)}$\\pm$ {round(kerr, 2)}')

plt.legend()
plt.show()
```

```python
tau = [2, 4, 10, 20, 30, 50]
tau = [1 / t / 100 * 1000 for t in tau]

nu  = [5.01, 2.45, 1.03, 0.46, 0.33, 0.19]

k, b, kerr, berr = LeastSquares(tau, nu)
```

```python
plt.figure()
plt.title("$\\delta \\nu (1/T)$")
plt.xlabel("$1/T \\cdot 10^{-3} c^{-1}$")
plt.ylabel("$\\delta \\nu, кГц$")

plt.grid(True, linestyle="--")
plt.errorbar([i for i in tau], [i for i in nu], fmt=".k")
x = np.array([min(tau), max(tau)])
plt.plot(x, k*x + b, label=f'k = {round(k, 2)} $\\pm$ {round(kerr, 2)}')

plt.legend()
plt.show()
```

```python
m = [10, 40, 60, 80, 90] # percent
m = [i / 100 for i in m]
a_a = [30.3/531.7,
98.9/531.7,
147.7/531.7,
195.3/531.7,
296.3/531.7]

k, b, kerr, berr = LeastSquares(m, a_a)
```

```python
plt.figure()
plt.title("$a_{\\text{бок}}/a_{\\text{осн}}(m)$")
plt.xlabel("$m, \\%$")
plt.ylabel("$a_{\\text{бок}}/a_{\\text{осн}}$")

plt.grid(True, linestyle="--")
plt.errorbar([i for i in m], [i for i in a_a], fmt=".k")
x = [i * 100 for i in x]
x = np.array([min(m), max(m)])

plt.plot(x, k*x + b, label=f'k $\\approx$ {round(k, 2)} $\\pm$ {round(kerr, 3)}')

plt.legend()
plt.show()
```
